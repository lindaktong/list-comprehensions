# List Comprenhensions 

```python
lst = []
for i in range(10):
    lst.append(i)
```

```python
lst
```

```python
lst = [i for i in range(10)]
```

```python
lst
```

Construct a list of all the number 1-1000 inclusive that are divisible by 8.

```python
lst = [i for i in range(1001) if i % 8 == 0]
```

```python
lst
```

```python
test = []
for i in range(1001):
    if i % 8 == 0:
        test.append(i)
```

```python
print(test)
```

Construct a list of all the numbers in 1-1000 inclusive that contain 3 as a digit at least once.

```python
lst = [i for i in range(1001) if str(3) in str(i)]
```

```python
print(lst)
```

```python
num_dict = {}
for i in range(1001):
    num_dict[i] = 1 + 1
```

```python
num_dict
```

```python
num_dict = {i : i + 1 for i in range(1001)}
```

```python
nums_to_strings = {str(i): i for i in range(1001)}
```

```python
nums_to_strings
```

```python
d = {}
for i in range(1001):
    lst = []
    for j in range(i, i + 10):
        lst.append(j)
    d[str(i)] = lst
print(d)
```
